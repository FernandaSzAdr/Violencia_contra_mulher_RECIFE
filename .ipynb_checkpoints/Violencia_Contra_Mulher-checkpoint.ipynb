### Andamento

```python
df_andamentos = pd.read_csv(local_andamentos)
```

```python
df_andamentos.head(10)
```

```python
df_andamentos[(df_andamentos['numero'] == 132120188170001)]
```

```python
numeros_aux = []
data_inicio_aux = []
data_fim_aux = []
duracao_aux = []

for numero in df_andamentos['numero'].unique():
    lista = list(df_andamentos[(df_andamentos['numero'] == int(numero))]['data'])
    lista = [datetime.strptime(lista[i], "%d/%m/%Y %H:%M:%S") for i in range(len(lista))]
    
    duracao = abs(lista[0] - lista[-1]).days

    numeros_aux.append(numero)
    data_inicio_aux.append(min(lista))
    data_fim_aux.append(max(lista))
    duracao_aux.append(duracao)
```

```python
duracao_andamentos = pd.DataFrame({
    'Numero': numeros_aux,
    'Data Inicio': data_inicio_aux,
    'Data Fim': data_fim_aux,
    'Duracao (em dias)': duracao_aux
})
```

```python
duracao_andamentos.sort_values(by='Duracao (em dias)')
```
